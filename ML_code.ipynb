Increase brightness and rotate image by 90 degree.

```python
import numpy as np
import matplotlib.pyplot as plt
m=np.array([[10,20,30],[40,50,60],[70,80,90]])
b=m+10
t=b.T
r=np.flip(t,axis=1)
fig,axes=plt.subplots(1,4,figsize=(16,4))
axes[0].imshow(m,cmap='Purples',vmin=0,vmax=100)
axes[0].set_title("Original image(Matrix)")
axes[0].axis('off')

axes[1].imshow(b,cmap='Purples',vmin=0,vmax=100)
axes[1].set_title("Brightness Adjusted")
axes[1].axis('off')

axes[2].imshow(t,cmap='Purples',vmin=0,vmax=100)
axes[2].set_title("Transposed image")
axes[2].axis('off')

axes[3].imshow(r,cmap='Purples',vmin=0,vmax=100)
axes[3].set_title("Rotated image(90)")
axes[3].axis('off')

plt.tight_layout()
plt.show()
```

Weather data analysis using Heatmaps and graphs

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

t=np.array([
    [20,22,21,23,25,24,22],
    [15,17,18,20,19,21,16],
    [30,28,31,33,32,35,34]
    ])

plt.figure(figsize=(8,5))
sns.heatmap(t,annot=True,cmap="coolwarm",xticklabels=["Mon","Tue","Wed","Thurs","Fri","Sat","Sun"],yticklabels=["City A","City B","City C"],cbar=True)
plt.title("Weekly Temp Heatmap")
plt.xlabel("Days")
plt.ylabel("Cities")
plt.show()

days=["Mon","Tue","Wed","Thurs","Fri","Sat","Sun"]
plt.figure(figsize=(8,5))
plt.plot(days,t[0],label="City A",marker='o')
plt.plot(days,t[1],label="City B",marker='o')
plt.plot(days,t[2],label="City C",marker='o')
plt.title("Weekly Temp Line Graph")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```
